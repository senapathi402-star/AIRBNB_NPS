**What is NPS?**

The term "NSP" is probably a typo of Net Promoter Score (NPS), a popular statistic used to measure customer satisfaction and loyalty that asks consumers on a scale of 0 to 10 how likely they are to suggest a company's goods or services.  Businesses may use the score as a single, comparable figure to gauge customer impression and monitor progress over time.

**Categorizing Responses:**
*  Promoters: Customers who score 9 or 10.
*  Passives: Customers who score 7 or 8.
*  Detractors: Customers who score 0 to 6.

**The total number of Promoters:**

Percentage of total number of reviews in a score > 8 must be calculated in order to obtain the Net Promoters score.

```python
arr[(arr>8)].shape
```

```python
promoters = round((np.sum(arr>8) / arr.size) * 100,2)
print(f'{promoters}%')
```

*   A total of 609 customers showed satisfaction with the ARIBNB's services.
*   The percentage of the total promoters is 52.19% of all over the total    reviews


**The total number of Detractors:**

Percentage total number of reviews in a score < 7 must be calculated in order to obtain the Net Promoters score.

```python
arr[(arr<7)].shape
```

```python
detractors = round((np.sum(arr<7) / arr.size) * 100,2)
print(f'{detractors}%')
```

*   A total of 332 customers showed dissatisfaction with the ARIBNB's services.
*   percentage of total detractors are 28.45%



**Net Promoters Score:**

download.png

```python
nps = promoters-detractors
print(f'{nps}%')
```

* The result of the Net Promoter Score is 23.74%
